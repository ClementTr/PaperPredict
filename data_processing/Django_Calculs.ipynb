```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv("data_2014_2016.csv", sep=";")
df1.head()
```

```python
df2 = pd.read_csv("df2_cleaned_full.csv")
df2.head()
```

```python
df = df1.merge(df2, how='left', on="Kiosque")
df["Date"] =  pd.to_datetime(df["Date"], format='%Y-%m-%d', errors='coerce') 
df.set_index(['Date'], inplace=True)
df["Year"] = df.index.year
df.head()
```

# DEPARTEMENT - Dept CSV

```python
def change_date_js(text):
    text = str(text)
    year = text[:4]
    month = text[5:7]
    return month + "-" + year

def create_dept_csv(dept):
    df_dept = df[df["code_dept"] == dept]
    df_dept["Sales"].resample("m").sum()
    df_dept["Delivered"].resample("m").sum()
    df_dept_months = pd.DataFrame({"Sales": df_dept["Sales"].resample("m").sum(), 
                               "Delivered": df_dept["Delivered"].resample("m").sum()})
    df_dept_months.reset_index(inplace=True)
    df_dept_months["Date"] = df_dept_months["Date"].apply(lambda x: change_date_js(x))
    df_dept_months.to_csv("dept_csv/df_dept_" + dept[1:] + "_months.csv", index=False)
    

for code_dept in np.unique(df["code_dept"]).ravel():
    create_dept_csv(code_dept)
```

```python
df_list = []

for i in range(1,96):
    df_temp = pd.read_csv("dept_csv/df_dept_" + str(i).zfill(2) + "_months.csv")
    df_temp["code_dept"] = str(i).zfill(2)
    df_list.append(df_temp)
    
df_dept = pd.concat(df_list)
df_dept.to_csv("df_dept_sales_deliveries.csv", index=False)
print(df_dept.shape)
df_dept.head()
```

# DEPARTEMENT - Dict dept Stats

```python
df.head()
```

```python
df_dept_stats_ind = df[["Delivered", "Sales", "code_dept", "Year"]].groupby(["code_dept", "Year"])\
                    .agg({'Sales': ['min', 'max', 'mean'], 'Delivered': ['min', 'max', 'mean'] }).copy()
df_dept_stats_ind.head()
```

```python
df_dept_stats = df_dept_stats_ind.copy()
df_dept_stats["Sales_min"] = df_dept_stats["Sales"]["min"]
df_dept_stats["Sales_max"] = df_dept_stats["Sales"]["max"]
df_dept_stats["Sales_mean"] = df_dept_stats["Sales"]["mean"]
df_dept_stats["Delivered_min"] = df_dept_stats["Delivered"]["min"]
df_dept_stats["Delivered_max"] = df_dept_stats["Delivered"]["max"]
df_dept_stats["Delivered_mean"] = df_dept_stats["Delivered"]["mean"]
del df_dept_stats["Sales"]
del df_dept_stats["Delivered"]
df_dept_stats.reset_index(inplace=True)
df_dept_stats["code_dept"] = df_dept_stats["code_dept"].str[1:]
df_dept_stats.head()
```

```python
import pickle

col = ['Sales_min', 'Sales_max', 'Sales_mean', 'Delivered_min', 'Delivered_max', 'Delivered_mean']

for c in col:
    my_dict = df_dept_stats.round(2).pivot(index='Year', columns='code_dept', values=c).to_dict()
    name = "dict_dept_" + str(c) + '.pickle'
    with open(name, 'wb') as handle:
        pickle.dump(my_dict, handle, protocol=pickle.HIGHEST_PROTOCOL)
    """
    with open('filename.pickle', 'rb') as handle:
        b = pickle.load(handle)"""
```

```python
path = "../newspaper/static/newspaper/data/df_kioques_villes.csv"
df_init = pd.read_csv(path)
df_init.head()
```

```python
df_2016 = df[df["Year"] == 2016].copy()
df_2016["Ruptures"] = np.where(df_2016.Delivered == df_2016.Sales, 1, 0)
df_dept_rupt = df_2016[["Ruptures", "Kiosque", "Ville", "Implantation", "Segment"]].groupby(["Kiosque", "Ville", "Implantation", "Segment"])\
                    .agg({'Ruptures': ['sum', 'min', 'max', 'mean'] }).copy()
df_dept_rupt.head()
```

```python
df_dept_rupt_2016 = df_dept_rupt.copy()
df_dept_rupt_2016["Ruptures_sum"] = df_dept_rupt_2016["Ruptures"]["sum"]
df_dept_rupt_2016["Ruptures_mean"] = df_dept_rupt_2016["Ruptures"]["mean"]
df_dept_rupt_2016["Ruptures_min"] = df_dept_rupt_2016["Ruptures"]["min"]
df_dept_rupt_2016["Ruptures_max"] = df_dept_rupt_2016["Ruptures"]["max"]
del df_dept_rupt_2016["Ruptures"]
df_dept_rupt_2016.reset_index(inplace=True)
df_dept_rupt_2016.head()
```

```python
df_dept_rupt_2016["Gravity"] = np.where(df_dept_rupt_2016["Ruptures_mean"] > df_dept_rupt_2016["Ruptures_mean"].quantile(0.66), 1,
                                       np.where(df_dept_rupt_2016["Ruptures_mean"] > df_dept_rupt_2016["Ruptures_mean"].quantile(0.33), 0, -1))
print(df_dept_rupt_2016["Gravity"].value_counts())
df_dept_rupt_2016["Implantation"] = df_dept_rupt_2016["Implantation"].apply(lambda x: x.title())
df_dept_rupt_2016["Segment"] = df_dept_rupt_2016["Segment"].apply(lambda x: x.title())
df_dept_rupt_2016["Ruptures_mean"] = df_dept_rupt_2016["Ruptures_mean"].apply(lambda x: str(round(x*100))+"%")

df_dept_rupt_2016.head()
```
